```python
# Required Libs
# Libs for Modeling and Matrices
import numpy as np
import pandas as pd

# Libs for graphical analysis
import matplotlib.pyplot as plt
import seaborn as sns

# Lib to ignore warnings
import warnings

# Disabling warnings
warnings.filterwarnings('ignore')
```

```python
# Loading the file
from google.colab import files
upload = files.upload()
```

```python
# Reading the data
Database = pd.read_csv('Salary_Data.csv')

# Check
Database.head()
```

```python
# Dimension
Database.shape
```

```python
# Null fields
Database.isnull().sum()
```

```python
# Null Chart
sns.heatmap( Database.isnull(), cbar = False );
```

```python
Database.describe()
```

```python
plt.figure( figsize = (10, 6) )
plt.title('Income Analysis')
sns.kdeplot( Database['Salary'], color = 'green' );
```

```python
plt.figure( figsize = (10, 6) )
plt.title('Analysis Years Experience')
sns.kdeplot( Database['YearsExperience'], color = 'blue' );
```

```python
# Distribution
plt.figure( figsize = (10, 6) )
plt.title('Income Bloxplot')
sns.boxplot( Database['Salary'] );
```

```python
# Distribution
plt.figure( figsize = (10, 6) )
plt.title('Years Experience Bloxplot')
sns.boxplot( Database['YearsExperience'] );
```

```python
# Distribution
plt.figure( figsize = (10, 6) )
sns.scatterplot( data = Database, x = 'Salary', y = 'YearsExperience' );
```

```python
# Distribution
plt.figure( figsize = (10, 6) )
sns.regplot( data = Database, x = 'Salary', y = 'YearsExperience' );
```

```python
# Correlation
Correlation = Database.corr('pearson')
plt.figure( figsize = (10, 6) )
sns.heatmap( Correlation, annot = True, cbar = True );
```

```python
# Correlation
Correlation = Database.corr('spearman')
plt.figure( figsize = (10, 6) )
sns.heatmap( Correlation, annot = True, cbar = True  );
```
